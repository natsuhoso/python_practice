```python
import numpy as np
```

```python
def subplots():
    plt.figure(figsize=(12, 8))
    plt.subplot(2, 2, 1)
    plt.xticks(())
    plt.yticks(())
    plt.text(0, 1, 'subplot(2,2,1)', ha='center', va='center',size=20, alpha=1)

    plt.subplot(2, 4, 3)
    plt.xticks(())
    plt.yticks(())
    plt.text(0, 0.5, 'subplot(2,2,2)', ha='left', va='center',size=20, alpha=.5)
    
    plt.subplot(2, 4, 4)
    plt.xticks(())
    plt.yticks(())

    ###############
    plt.subplot(2, 2, 3)
    X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
    C, S = np.cos(X), np.sin(X)

    plt.plot(X, C, color="red", linewidth=1.0, linestyle="-", label="cosine")
    plt.plot(X, S, color="blue", linewidth=1.0, linestyle="-", label="sine")

    plt.xlim(X.min() * 1.1, X.max() * 1.1)
    plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])#latex
    
    plt.ylim(C.min() * 1.1, C.max() * 1.1)
    #plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
    plt.yticks([-1, 0, +1],[r'$-1$','', r'$+1$'])
    
    plt.legend(loc='upper left')
    
    plt.fill_between(X, 0, C, C < 0, color='red', alpha=.25)

    ax = plt.gca()  # gca stands for 'get current axis'
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    t = 2 * np.pi / 3
    plt.plot([t, t], [0, np.sin(t)], color='blue', linewidth=2, linestyle="-.")
    plt.scatter([t, ], [np.sin(t), ], 50, color='blue')
    
    plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{1}{\sqrt{3}}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(20, 10), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    #######
    
    plt.subplot(2, 2, 4)
    plt.xlim(0, 10)
    plt.xticks(())
    plt.yticks(())
    ax = plt.gca()
    ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))
    ax.xaxis.set_major_locator(plt.FixedLocator([0, 2, 8, 9, 10]))

    plt.tight_layout()
    plt.show()
```

```python
subplots()
```

```python
def piechart():
    n = 20
    Z = np.ones(n)
    Z[0] = 3
    Z[-1] *= 2

    plt.axes([0.025, 0.025, 0.95, 0.95])

    #plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])
    plt.pie(Z, explode=np.ones(n)*.05, colors = ['%f' % (i/float(n)) for i in range(n)])
    plt.axis('equal')
    plt.xticks(())
    plt.yticks()

    plt.show()
```

```python
piechart()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
def threeD():
    fig = plt.figure()
    ax = Axes3D(fig)
    X = np.arange(-4, 4, 0.25)
    Y = np.arange(-4, 4, 0.25)
    X, Y = np.meshgrid(X, Y)
    R = np.sqrt(X ** 2 + Y ** 2)
    Z = np.sin(R)

    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
    ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
    ax.set_zlim(-2, 2)

    plt.show()
```

```python
threeD()
```
