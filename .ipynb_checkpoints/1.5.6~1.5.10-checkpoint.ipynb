### 1.5.8 FFT

```python
time_vec = np.arange(0, 20, 0.001)
```

```python
signal = np.sin(2 * np.pi / 5 * time_vec) + 0.5 * np.random.randn(time_vec.size)
```

```python
plt.plot(time_vec, signal)
```

```python
from scipy import fftpack
```

```python
sigfft = fftpack.fft(signal)
```

```python
power = np.abs(sigfft)
```

```python
sample_freq = fftpack.fftfreq(signal.size, d=0.001)
```

```python
pos_mask = np.where(sample_freq > 0)
```

```python
freqs = sample_freq[pos_mask]
```

```python
peak_freq = freqs[power[pos_mask].argmax()]
```

```python
high_freq_fft = sigfft.copy()
```

```python
high_freq_fft[np.abs(sample_freq) > peak_freq] = 0
```

```python
filtered_sig = fftpack.ifft(high_freq_fft)
```

```python
plt.plot(time_vec, filtered_sig)
```

```python
def plotfft():
    plt.plot(time_vec, signal)
    plt.plot(time_vec, filtered_sig)
    plt.show()
```

```python
plotfft()
```

```python
im_new = fftpack.ifft2(imf).real
```

## 1.5.9 signal processing

```python
from scipy import signal
```

```python
t = np.linspace(0, 5, 100)
```

```python
x = np.sin(t)
```

```python
x_resampled = signal.resample(x, 25)
```

```python
plt.plot(t, x)
plt.plot(t[::4], x_resampled, 'o')
```

detrending : remove linear trend from signal

```python
t = np.linspace(0, 5, 100)
x = t + np.random.normal(size=100)
x_detrended = signal.detrend(x)
plt.plot(t, x)
plt.plot(t, x_detrended)
```

```python
time_step = .01
time_vec = np.arange(0, 70, time_step)
sig = np.sin(0.5 * np.pi * time_vec * (1 + .1 * time_vec))
freqs, times, spectrogram = signal.spectrogram(sig)
freqs, psd = signal.welch(sig)

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.plot(time_vec, sig)

plt.subplot(1, 3, 2)
plt.imshow(spectrogram, aspect='auto', cmap='hot_r', origin='lower')

plt.subplot(1, 3, 3)
plt.semilogx(freqs, psd)
```
